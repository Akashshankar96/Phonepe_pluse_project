# **Aggregated Transaction**

```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yrs=os.listdir(p_i)    
    for j in Agg_yrs:
        p_j=p_i+j+"/"
        Agg_yrs_list=os.listdir(p_j)        
        for k in Agg_yrs_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
pd.DataFrame(clm)

```

**Aggregated Users**

```python
import os
import json
import pandas as pd

# This is to direct the path to get the data as states
path = "/content/pulse/data/aggregated/user/country/india/state/"
Agg_state_user_list = os.listdir(path)

# This is to extract the data's to create a dataframe
clm = {'State':[], 'Year':[], 'Quarter':[], 'registeredUsers':[], 'App_open_count':[], 'Transaction_amount':[]}
for i in Agg_state_user_list:
    p_i = path + i + "/"
    Agg_yrs = os.listdir(p_i)    
    for j in Agg_yrs:
        p_j = p_i + j + "/"
        Agg_yrs_list = os.listdir(p_j)        
        for k in Agg_yrs_list:
            p_k = p_j + k
            with open(p_k, 'r') as f:
                data = json.load(f)
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quarter'].append(int(k.strip('.json')))
                clm['registeredUsers'].append(data['data']['aggregated']['registeredUsers'])
                clm['App_open_count'].append(data['data']['aggregated']['appOpens'])
                try:
                    clm['Transaction_amount'].append(data['data']['usersByDevice'][0]['amount'])
                except:
                    clm['Transaction_amount'].append(None)

# Successfully created a dataframe
df1 = pd.DataFrame(clm)
df1
```

# **MAP transaction**

```python
import sqlite3
import os
import json
import pandas as pd

# This is to direct the path to get the data as states
path = "/content/pulse/data/map/transaction/hover/country/india/state/"
map_state_list = os.listdir(path)

# This is to extract the data's to create a dataframe
clm = {'State': [],'District': [],'Year': [], 'Quarter': [], 'Metric_type': [], 'Metric_count': [], 'Metric_amount': []}
for i in map_state_list:
    p_i = path + i + "/"
    Agg_yrs = os.listdir(p_i)
    for j in Agg_yrs:
        p_j = p_i + j + "/"
        Agg_yrs_list = os.listdir(p_j)
        for k in Agg_yrs_list:
            p_k = p_j + k
            with open(p_k, 'r') as f:
                data = json.load(f)
                clm['District'].append(data['data']['hoverDataList'][0]['name'])
                clm['State'].append(i)
                clm['Quarter'].append(j)
                clm['Year'].append(int(k.strip('.json')))
                for metric in data['data']['hoverDataList'][0]['metric']:
                    clm['Metric_type'].append(metric['type'])
                    clm['Metric_count'].append(metric['count'])
                    clm['Metric_amount'].append(metric['amount'])

df3 = pd.DataFrame(clm)

df3
```

```python
import sqlite3

conn = sqlite3.connect("Transaction.db", timeout=10)
s = conn.cursor()

# Create a table in database
s.execute('''CREATE TABLE IF NOT EXISTS Map_Transaction(
            State TEXT,
            District TEXT,
            Year TEXT,
            Quarter TEXT,
            Metric_type TEXT,
            Metric_count TEXT,
            Metric_amount TEXT
            )''')

# Insert the data into the table
s.executemany('INSERT INTO Map_Transaction VALUES(?,?,?,?,?,?,?)',
              zip(clm['State'], clm['District'], clm['Year'], clm['Quarter'], clm['Metric_type'], clm['Metric_count'], clm['Metric_amount']))

# Commit changes and close connection
conn.commit()
conn.close()

```

# **TOP Transaction**

```python
import os
import json
import pandas as pd
import sqlite3

# set the directory path
path = '/content/pulse/data/top/transaction/country/india/state'

# create a dictionary to store the extracted data
clm = {'Entity_type': [], 'Entity_name': [], 'Year': [], 'Quarter': [], 'Total_transactions': [], 'Total_value': []}

# loop over the directory and extract the data
for state in os.listdir(path):
    state_path = os.path.join(path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for quarter in os.listdir(year_path):
            quarter_path = os.path.join(year_path, quarter)
            with open(quarter_path, 'r') as f:
                data = json.load(f)
                if data is not None and 'data' in data:
                    for entity_type, entity_data in data['data'].items():
                        if entity_data is not None:
                            for entity in entity_data:
                                clm['Entity_type'].append(entity_type)
                                clm['Entity_name'].append(entity['entityName'])
                                clm['Year'].append(year)
                                clm['Quarter'].append(int(quarter.strip('.json')))
                                clm['Total_transactions'].append(entity['metric']['count'])
                                clm['Total_value'].append(entity['metric']['amount'])

# create a pandas dataframe from the extracted data
df5 = pd.DataFrame(clm)

# create a connection to the SQLite database
conn = sqlite3.connect('map_data.db')

# insert the data into the database
df5.to_sql('top_transactions', conn, if_exists='replace', index=False)

# commit changes and close the connection
conn.commit()
conn.close()
```
